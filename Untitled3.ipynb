```python
# Funktion zum Berechnen des Preises
def berechne_preis(artikel, menge):

    # Preisliste als Dictionary
    preise = {
        "Kaffee": 1.50,
        "Tee": 1.20,
        "Wasser": 0.80,
        "Snack": 2.00
    }

    # Prüfen ob Artikel existiert
    if artikel not in preise:
        print(f"Fehler: {artikel} ist unbekannt.")
        return None

    # Gesamtpreis berechnen
    return preise[artikel] * menge



# Klasse Tageskasse
class Tageskasse:

    # Konstruktor
    def __init__(self):

        # Gesamtumsatz
        self.umsatz = 0.0

        # Liste aller Bestellungen
        self.bestellungen = []


    # Bestellung hinzufügen
    def bestellen(self, artikel, menge):

        # Preis berechnen
        betrag = berechne_preis(artikel, menge)

        # Fehlerfall prüfen
        if betrag is None:
            return

        # Bestellung speichern
        bestellung = {
            "artikel": artikel,
            "menge": menge,
            "betrag": betrag
        }

        self.bestellungen.append(bestellung)

        # Umsatz erhöhen
        self.umsatz += betrag

        print(f"{menge}x {artikel} hinzugefügt - {betrag:.2f} €")


    # Tagesübersicht
    def tagesabrechnung(self):

        print("\n----- Tagesabrechnung -----")

        # Alle Bestellungen anzeigen
        for bestellung in self.bestellungen:

            print(
                f"{bestellung['menge']}x "
                f"{bestellung['artikel']} "
                f"= {bestellung['betrag']:.2f} €"
            )

        print("---------------------------")
        print(f"Gesamtumsatz: {self.umsatz:.2f} €")



# Bestellliste
bestellungen_des_tages = [
    ("Kaffee", 3),
    ("Snack", 2),
    ("Tee", 5),
    ("Pizza", 1),   # unbekannter Artikel
    ("Wasser", 4),
    ("Kaffee", 7),
]


# Tageskasse erstellen
kasse = Tageskasse()


# Schleife durch alle Bestellungen
for artikel, menge in bestellungen_des_tages:

    kasse.bestellen(artikel, menge)


# Tagesabrechnung anzeigen
kasse.tagesabrechnung()
```
