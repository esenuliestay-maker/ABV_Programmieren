```python
def berechne_endpreis(preis):
    if preis >= 100:
        rabatt = 0.15
    elif preis >= 50:
        rabatt = 0.10
    elif preis >= 20:
        rabatt = 0.05
    else:
        rabatt = 0

    return preis * (1 - rabatt)


for p in [15.00, 34.50, 72.90, 120.00]:

    if p >= 100:
        rabatt = 15
    elif p >= 50:
        rabatt = 10
    elif p >= 20:
        rabatt = 5
    else:
        rabatt = 0

    endpreis = berechne_endpreis(p)

    print("Original Preis:", p,
          "Rabatt:", rabatt, "%",
          "Endpreis:", endpreis, "€")
```
