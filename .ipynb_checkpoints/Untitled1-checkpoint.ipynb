```python
profit = 5000000
print(profit)

bonus = 400000
print(bonus)

Josephine = profit * 0.3
print(Josephine)

Otto = profit * 0.25
print(Otto)

Rest = profit - Josephine - Otto
print(Rest)

Paul = Rest / 2
print(Paul)

Yestay = Rest / 2 + bonus
print(Yestay)

print(f"Josephine: {Josephine} $")

print(f"Josephine: {Josephine} € | Otto: {Otto} € | Paul: {Paul} € | Yestay: {Yestay} €")

```
