```python
class Kurs:

    # Konstruktor: wird automatisch ausgeführt,
    # wenn ein Objekt erstellt wird
    def __init__(self, titel, dozent, max_teilnehmer):

        # Attribute speichern
        self.titel = titel
        self.dozent = dozent
        self.max_teilnehmer = max_teilnehmer

        # Leere Liste für Teilnehmende
        self.teilnehmer = []


    # Methode zum Anmelden
    def anmelden(self, name):

        # Prüfen ob Kurs voll ist
        if len(self.teilnehmer) >= self.max_teilnehmer:
            print(f"Kurs voll - {name} konnte nicht angemeldet werden.")

        else:
            self.teilnehmer.append(name)
            print(f"{name} wurde erfolgreich angemeldet.")


    # Methode zum Abmelden
    def abmelden(self, name):

        # Prüfen ob Person in der Liste ist
        if name in self.teilnehmer:
            self.teilnehmer.remove(name)
            print(f"{name} wurde erfolgreich abgemeldet.")

        else:
            print(f"{name} ist nicht angemeldet.")


    # Übersicht anzeigen
    def display(self):

        print("----- Kursübersicht -----")
        print(f"Titel: {self.titel}")
        print(f"Dozent: {self.dozent}")
        print(f"Teilnehmer: {self.teilnehmer}")
        print(f"Belegung: {len(self.teilnehmer)}/{self.max_teilnehmer}")
```

```python
bwl_grundlagen = Kurs("BWL Grundlagen", "Prof. Dr. Richter", 3)

bwl_grundlagen.anmelden("Anna Bauer")
bwl_grundlagen.anmelden("Ben Schulz")
bwl_grundlagen.anmelden("Clara Hoffmann")

# Kurs voll
bwl_grundlagen.anmelden("Daniel Meier")

# Ben entfernen
bwl_grundlagen.abmelden("Ben Schulz")

# Nicht vorhanden
bwl_grundlagen.abmelden("Felix Wagner")

# Übersicht anzeigen
bwl_grundlagen.display()
```
